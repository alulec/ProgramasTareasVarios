```
"""
a, b, c y d son las constantes de x^2, x y la constante respectivamente

e y f son las unidades de inicio, final y seguimiento del uso del for
"""
a = float(input("Dame el coeficiente cubico: "))
b = float(input("Dame el coefciente cuadrado: "))
c = float(input("Dame el coeficiente lineal: "))
d = float(input("Deme el valor del   Ter. independiente: "))
men = int(input("Deme el valor menor del intervalo: "))
may = int(input("Deme el valor mayor del intervalo: "))
noIter = int(input("Deme el # de iteraciones: "))

def f(x):
     h1 = (a*(x**3)) + (b*(x**2)) + (c*x) + (d)
     return h1

#print(f(x))

def bisection_method(cota_izq,cota_der,iterations):
    rang = (cota_izq, cota_der)
    midpoint = (rang[0] + rang[1])/2


    for i in range(iterations):
        f1 = f(rang[0])
        f2 = f(rang[1])
        fm = f(midpoint)
        print("Iteración ",i)
        print(f"Intervalo: ({e},{f}), Punto medio: {midpoint}")
        print(f"f(x) en el intervalo: Izquierda {f1},  Centro {fm}, Derecha {f2}")


        if (fm==0):
            return midpoint
        else:
            if (f1*fm)<0 :
                print("Elijo el intervalo izquierdo ‹(ᵒᴥᵒ<)\n")
                rang = (rang[0], midpoint)
            else:
                print("Elijo el intervalo derecho (>ᵒᴥᵒ)>\n")
                rang = (midpoint, rang[1])

            midpoint = (rang[0] + rang[1])/2

    return midpoint

x = bisection_method(men, may, noIter)
print("La raiz x esta en: ",x)
print("f(x)=",f(x))
```
